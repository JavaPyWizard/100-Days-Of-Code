# Create a pie chart showing the market share of mobile OS (Android, iOS, others). Then recreate the same using a horizontal bar chart and observe which is easier to understand.

```python
import matplotlib.pyplot as plt

# Data
labels = ['Android', 'iOS', 'Others']
sizes = [71, 28, 1]
colors = ['#34A853', '#4285F4', '#AAAAAA']
explode = [0.1,0,0]

# Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, explode=explode, colors=colors, autopct='%1.1f%%', startangle=0)
plt.title('Mobile OS Market Share - 2025')
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Android', 'iOS', 'Others']
sizes = [71, 28, 1]
colors = ['#34A853', '#4285F4', '#AAAAAA']

# Plotting
plt.figure(figsize=(8, 4))
bars = plt.barh(labels, sizes, color=colors)

# Add labels to bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 1, bar.get_y() + bar.get_height() / 2,
             f'{width}%', va='center', fontsize=10)

plt.xlabel('Market Share (%)')
plt.title('Mobile OS Market Share - 2025')
plt.xlim(0, 100)
plt.tight_layout()
plt.show()
```
